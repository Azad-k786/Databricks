```
array_appli=[
             ('Raja',['TV','Refrigerator','Oven','AC']),
            ('Rajesh',['AC','WashingMachine',None]),
            ('Ramesh',['Grander','TV']),
            ('Rajesh',None)]

df_array=spark.createDataFrame(data=array_appli,schema=['Name','Application'])
df_array.printSchema()
display(df_array)
```

```
map_value=[
    ('Raja',{'TV':'LG','Refrigerator':'Samsung','Oven':'Phillipse','AC':'Voltas'}),
     ('Rajesh',{'AC':'Samsung','WashingMachine':'LG'}),
    ('Ramesh',{'Grander':'Preethi'}),
    ('Rajesh',None)]
df_map=spark.createDataFrame(data=map_value,schema=['Name','Brand'])
df_map.printSchema()
display(df_map)
```

# Explode Array Field

```
from pyspark.sql.functions import explode
df_array2=df_array.select(df_array.Name,explode(df_array.Application))
display(df_array2)
```

#Explode Map Field

```
df_map2=df_map.select(df_map.Name,explode(df_map.Brand))
display(df_map2)
```

#Explode_outer

```
from pyspark.sql.functions import explode_outer
df_array2=df_array.select(df_array.Name,explode_outer(df_array.Application))
display(df_array2)
```

# posexplode Array Feild

```
from pyspark.sql.functions import posexplode
df_array2=df_array.select(df_array.Name,posexplode(df_array.Application))
```
