```
Attendance=spark.read.parquet("/Bronze_Layer/AttendanceFact")
display(Attendance)
```

```
Attendance=Attendance.withColumn("Employee_Name_Null",f.when(Attendance.EmpName.isNull(),"Name is Null").otherwise(""))
Attendance=Attendance.withColumn("Project_Name",f.when(Attendance.ProjectName.isNull(),"Project Name is Null").otherwise(""))
Attendance=Attendance.select("EmpName","ProjectName","Present","Halfday","Businessdate",f.concat_ws(' ',Attendance.Employee_Name_Null,Attendance.Project_Name)
                             .alias("Rejected_Reason"))

display(Attendance)

At=Attendance.filter(~Attendance.Rejected_Reason.like(' '))

display(At)
```
