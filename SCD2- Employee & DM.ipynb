```
from pyspark.sql import functions as f

from pyspark.sql.functions import md5,concat_ws
```

```
Emp_df=spark.read.parquet("/Silver_Layer/Employee/Without_Null_record")
display(Emp_df)
```

```
df=Emp_df.select("EMP_FIRST_NAME","EMP_LAST_NAME","BILLABLE")
display(df)
```

```
Emp_df2=df.withColumn("Emp_Id", md5(concat_ws("||",*df.columns)))
display(Emp_df2)
```

```
%sql
drop table if exists em1;
create table em1
(
EMP_FIRST_NAME string,
EMP_LAST_NAME string,
BILLABLE string,
Emp_Id string
)
using delta
```

```
%sql
drop table if exists em2;
create table em2
(
EMP_FIRST_NAME string,
EMP_LAST_NAME string,
BILLABLE string,
Emp_Id string,
Start_date date,
End_date date
)
```
