```
from pyspark.sql import functions as f
from pyspark.sql.functions import md5,concat_ws
```

```
%sql 
create database GoldLayer comment 'In this Layer I used SCD-2 & craete Dim tabel '  location '/Gold_Layer/GoldLayer_db';
```

```
%sql
use GoldLayer
```

```
Emp_df=spark.read.parquet("/Silver_Layer/Employee/Without_Null_record")
```

```
Emp_df2=Emp_df.withColumnRenamed("LOCATION","Emp_LOCATION")
```

```
Emp_df3=Emp_df2.withColumn("Emp_Id", md5(concat_ws("||",*Emp_df2.columns)))
display(Emp_df3)
```

```
Emp_df3.write.format("delta").mode("overwrite").saveAsTable("SourceEmployee")
```

```
%sql
drop table if exists targetemp;
create table targetemp
(
EMP_FIRST_NAME string,
EMP_LAST_NAME string,
EMP_MIDDLE_NAME string,
JOINING_DATE date,
EMP_LOCATION string,
BILLABLE string,
IS_ACTV string,
LAST_WORKING_DATE date,
ADDRESS_LINE_1 string,
ADDRESS_LINE_2 string,
ADDRESS_LINE_3 string,
ADDRESS_LINE_4 string,
CITY string,
STATE string,
COUNTRY string,
ZIP int,
CONTACT int,
DEPTNAME string,
SUBDEPT_NAME string,
ACCOUNT_NUMBER int,
BANK_NAME string,
SALARY int,
DESIGNATION string,
Emp_ID string,
start_date date,
end_date date
)
USING delta
```

```
%sql
merge into targetemp a
using
(
  select Emp_ID as mergeKey, * from SourceEmployee
                                                                 
  union all
  
  select NULL as mergeKey, a.* from SourceEmployee a join targetemp b on a.Emp_ID = b.Emp_ID and b.EMP_FIRST_NAME != a.EMP_FIRST_NAME 
  )b 
  
on a.Emp_ID = b.mergeKey

when matched and b.EMP_FIRST_NAME != a.EMP_FIRST_NAME then  

  update set  a.End_Date = current_date()-1
  
  when not matched 
  
then insert (EMP_FIRST_NAME,EMP_LAST_NAME,EMP_MIDDLE_NAME,JOINING_DATE,EMP_LOCATION,BILLABLE,IS_ACTV,LAST_WORKING_DATE,
ADDRESS_LINE_1,ADDRESS_LINE_2,ADDRESS_LINE_3,ADDRESS_LINE_4,CITY,STATE,COUNTRY,ZIP,CONTACT,DEPTNAME,
SUBDEPT_NAME,ACCOUNT_NUMBER,BANK_NAME,SALARY,DESIGNATION,Emp_ID,start_date,end_date) values
(EMP_FIRST_NAME,EMP_LAST_NAME,EMP_MIDDLE_NAME,JOINING_DATE,EMP_LOCATION,BILLABLE,IS_ACTV,LAST_WORKING_DATE,
ADDRESS_LINE_1,ADDRESS_LINE_2,ADDRESS_LINE_3,ADDRESS_LINE_4,CITY,STATE,COUNTRY,ZIP,CONTACT,DEPTNAME,
SUBDEPT_NAME,ACCOUNT_NUMBER,BANK_NAME,SALARY,DESIGNATION,Emp_ID,current_date,'')
```

```
df=spark.table("targetemp")
display(df)
```

```
Employee_Dim=df.select("Emp_ID","EMP_FIRST_NAME","EMP_LAST_NAME","EMP_MIDDLE_NAME","start_date","end_date")
Address_Dim=df.select("Emp_ID","ADDRESS_LINE_1","ADDRESS_LINE_2","ADDRESS_LINE_3","ADDRESS_LINE_4","CITY","STATE","ZIP","start_date","end_date")
Salary_Dim=df.select("Emp_ID","ACCOUNT_NUMBER","BANK_NAME","start_date","end_date")
Department_Dim=df.select("Emp_ID","DEPTNAME","SUBDEPT_NAME","DESIGNATION","start_date","end_date")
```
