```
df=spark.read.format('csv').option("inferSchema",True).option("header",True).load("/FileStore/tables/emp.csv")
display(df)
```

```
df2=df.filter(df.salary<=5000)
display(df2)
```

```
df3=df.filter(df.gender.isNotNull())
display(df3)
```
