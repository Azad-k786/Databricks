```
%sql
create table Source_Em
(
EMP_FIRST_NAME string,
EMP_LAST_NAME string,
EMP_MIDDLE_NAME string,
JOINING_DATE date,
BILLABLE string,
IS_ACTV string,
LAST_WORKING_DATE date,
ADDRESS_LINE_1 string,
ADDRESS_LINE_2 string,
ADDRESS_LINE_3 string,
ADDRESS_LINE_4 string,
CITY string,
STATE string,
COUNTRY string,
ZIP int,
CONTACT int,
DEPTNAME string,
SUBDEPT_NAME string,
ACCOUNT_NUMBER int,
BANK_NAME string,
SALARY int,
DESIGNATION string,
Emp_ID string
)
USING delta
```
