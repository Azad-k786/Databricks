```
dept_data=(("HR",100),
           ("supply",200),
           ("sales",300),
           ("stocks",400)
          )
dept_data_Schema=("dept_name","dept_id")

dept_df=spark.createDataFrame(data=dept_data,schema=dept_data_Schema)
```

```
display(dept_df)
display(df)
```

#inner

```
df_join=dept_df.join(df,df.employee_dept_id == dept_df.dept_id)
```

```
display(df_join)
```

# Full Outer

```
df_full=df.join(dept_df,df.employee_dept_id==dept_df.dept_id,"fullouter")
```

```
display(df_full)
```

# leftOuter

```
df_left=df.join(dept_df,df.employee_dept_id==dept_df.dept_id,"left")
```

```
display(df_left)
```

```
df_semi=df.join(dept_df,df.employee_dept_id==dept_df.dept_id,"semi")
```

```
display(df_semi)
```

```
df_anti=df.join(dept_df,df.employee_dept_id==dept_df.dept_id,"anti")
display(df_anti)
```
