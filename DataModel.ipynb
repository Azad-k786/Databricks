```
%sql
MERGE INTO emp a
USING
(
  SELECT Emp_ID as mergeKey, * FROM Em
                                                                
  UNION ALL
  
  SELECT NULL as mergeKey, a.* FROM Em a JOIN emp b ON a.Emp_ID = b.Emp_ID AND b.EMP_FIRST_NAME != a.EMP_FIRST_NAME 
  )b 
  
ON a.Emp_ID = b.mergeKey

WHEN MATCHED AND b.EMP_FIRST_NAME != a.EMP_FIRST_NAME THEN  

  UPDATE SET     a.End_Date = current_timestamp()
  WHEN NOT MATCHED 
THEN INSERT (EMP_FIRST_NAME,EMP_MIDDLE_NAME,EMP_LAST_NAME,Emp_ID,Start_Date,End_Date) VALUES
(EMP_FIRST_NAME,EMP_MIDDLE_NAME,EMP_LAST_NAME,Emp_ID,current_timestamp(),'')
  
```
